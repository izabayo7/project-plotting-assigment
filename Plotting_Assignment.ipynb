```python
plt.hist(df['co2'], bins=10, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('CO2 Emissions (10 bins)')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel('Frequency')
```

```python
plt.hist(df['co2'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('CO2 Emissions (20 bins)')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel('Frequency')
```

```python
plt.hist(df['co2'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('CO2 Emissions (50 bins)')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel('Frequency')
```
